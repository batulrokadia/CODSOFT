```python
#importing the required libraries
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
df=pd.read_csv('/content/creditcard.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
df.isnull().sum()
```

There are no missing values in this dataset.

```python
df['Class'].value_counts()
```

This shows that the given dataset is highly unbalanced as more than 90% of the transactions are legal.

```python
legal=df[df.Class==0]
fraud=df[df.Class==1]
print(legal.shape)
print(fraud.shape)
```

```python
df.groupby('Class').mean()
```


Now, we will build a smaller dataset which will contain a balance of legal and fraud transactions.

Number of frauds are 492.

```python
legalsample=legal.sample(n=492)
```

Combining the fraud transactions with the sample of legal transactions

```python
balanced_dataset= pd.concat([legalsample, fraud], axis=0)
```

```python
balanced_dataset.shape
```

```python
balanced_dataset.head()
```

```python
balanced_dataset.tail()
```

```python
balanced_dataset['Class'].value_counts()
```

```python
balanced_dataset.groupby('Class').mean()
```

The means are similar to those of the original dataset so that means that our new dataset is quite similar to the original.

Now, we split the new dataset into feautures and target variable.

```python
x=balanced_dataset.drop(columns='Class', axis=1)
y=balanced_dataset['Class']
print(x)
print(y)
```

Splitting the new datset into training and testing data

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, stratify=y,random_state=2)
print(x.shape, x_train.shape, x_test.shape)
```

Now, we will train the model.

Generally, we use Logistic regression model for binary classification problems.

```python
model= LogisticRegression()
```

```python
#training the model with the training data
model.fit(x_train, y_train)
```

Now, we will evaluate the model's accuracy

```python
training_prediction=model.predict(x_train)
print("Accuracy of the model's prediction in percentage is :", (accuracy_score(y_train, training_prediction))*100, "%")
```

```python
testing_prediction=model.predict(x_test)
print("Accuracy of the model's prediction in percentage is :", (accuracy_score(y_test, testing_prediction))*100, "%")
```

```python
print(classification_report(testing_prediction, y_test))
```
